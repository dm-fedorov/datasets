Информация о данных:

Этот набор данных содержит 76 признаков. Но используются подмножества из 14 из них. В частности, дынне Кливленда - единственные данные, которые использовались в машинном обучении. 
Переменная "goal" показывает наличие сердечных заболеваний у пациента. Это целое число от 0 (отсутствие присутствия) до 4. Эксперименты с базой данных Cleveland были сосредоточены на простой попытке отличить присутствие (значения 1,2,3,4) от отсутствия (значение 0).

Имена и номера социального страхования пациентов были недавно удалены из данных, заменены на пустые значения. 

Файл Кливленда юыд обработан. Все четыре необработанных файла также существуют в этом каталоге. 

Посмотреть стоимость тестирования (пожертвование Питера Терни) можно в папке "расходы"

Полная версия:

1. id: идентификационный номер пациента 
2. ccf: номер социального страхования (Я заменил его фиктивным значением 0) 
3. age: возраст в годах 
4. sex: Пол (1 = мужчина; 0 = женщина) 
5. painloc: болит ли грудь (1 = в грудине; 0 = в противном случае) 
6. painexer (1 = провоцируется напряжением; 0 = в противном случае) 
7. relrest (1 = облегчение после отдыха; 0 = в противном случае) 
8. pncaden (сумма 5, 6 и 7) 
9. cp: тип боли в груди
    - Value 1: типичная стенокардия 
    - Value 2: атипичная стенокардия  
    - Value 3: боль не от стенокардии
    - Value 4: бессимптомная
10. trestbps: кровяное давление в состоянии отдыха (в мм рт. ст. при поступлении в больницу) 
11. htn 
12. chol: холестерин в мг/дл 
13. smoke:  1 = да; 0 = нет (или не курит) 
14. cigs: количество сигарет в день 
15. years (количество лет в качестве курильщика) 
16. fbs: (уровень сахара в крови натощак > 120 мг/дл) (1 = true; 0 = false) 
17. dm (1 = история диабета; 0 = Нет такой истории) 
18. famhist: семейная история ишемической болезни сердца (1 = да; 0 = Нет) 
19. restecg: электрокардиографические результаты в состоянии отдыха
    - Value 0: норма
    - Value 1: наличие аномалии волны ST-T (инверсии волны T и/или высота или депрессия ST > 0,05 МВ)  
    - Value 2: показывает вероятную или определенную гипертрофию левого желудочка по критериям Эстеса 
20. ekgmo (месяц проведения ЭКГ) 
21. ekgday(день проведения ЭКГ) 
22. ekgyr (год проведения ЭКГ) 
23. dig (применялся ли дигиталис при ЭКГ: 1 = да; 0 = нет) 
24. prop (бета блокатор, используемый во время ЭКГ: 1 = да; 0 = нет) 
25. nitr (нитраты, используемые при ЭКГ: 1 = да; 0 = Нет: 1 = да; 0 = нет) 
26. pro (блокатор кальциевых каналов, используемый во время ЭКГ: 1 = да; 0 = нет) 
27. diuretic (мочегонное средство, используемое при ЭКГ: 1 = да; 0 = нет) 
28. proto: протокол исследовани
    - 1 = Bruce 
    - 2 = Kottus 
    - 3 = McHenry 
    - 4 = fast Balke 
    - 5 = Balke 
    - 6 = Noughton 
    - 7 = bike 150 kpa min/min (Not sure if "kpa min/min" is what was written!) 
    - 8 = bike 125 kpa min/min 
    - 9 = bike 100 kpa min/min 
    - 10 = bike 75 kpa min/min 
    - 11 = bike 50 kpa min/min 
    - 12 = arm ergometer
29. thaldur: продолжительность теста с физической нагрузкой в течение нескольких минут 
30. thaltime: время, когда ST мера снизилась 
31. met: достигнутая mets  
32. thalach: максимальная достигнутая ЧСС 
33. thalrest: ЧСС  
34. tpeakbps: пиковое кровяное давление (первая из 2 частей) 
35. tpeakbpd: пиковое кровяное давление (во второй 2 частей)
36. пустой
37. trestbpd: кровяное давление покоя 
38. exang: стенокардия, вызванная физической нагрузкой (1 = да; 0 = нет) 
39. xhypo: (1 = да; 0 = нет) 
40. oldpeak = ST снижение, вызванное физическими упражнениями (относительно отдыха)
41. slope: отклонение пика ST:
    - Value 1: вверз
    - Value 2: ровно
    - Value 3: вниз
42. rldv5: высота в покое 
43. rldv5e: высота на пике 
44. ca: количество крупных копиляров (0-3) на рентгене
45. restckm: неактуально 
46. exerckm: неактуально 
47. restef: rest radionuclides (sp?) ejection fraction
48. restwm: rest wall (sp?) motion abnormality  
    - 0 = Нет 
    - 1 = легкая или умеренная 
    - 2 = умеренная или тяжелая 
    - 3 = akinesis or dyskmem (sp?) 
49. exeref: exercise radinalid (sp?) ejection fraction
50. exerwm: exercise wall (sp?) motion 
51. thal: 3 = нормальный; 6 = фиксированный дефект; 7 = реверсивный дефект 
52. thalsev: не используется
53. thalpul: не используется 
54. earlobe: не используется 
55. cmo: месяц сердечной катетеризации 
56. cday: день сердечной катетеризации
57. cyr: год сердечной катетеризации
58. num: диагностики сердечных заболеваний (ангиография) 
    - Value 0: сужение диаметра < 50% 
    - Value 1: сужение диаметра > 50%
59. lmt 
60. ladprox 
61. laddist 
62. diag 
63. cxmain 
64. ramus 
65. om1 
66. om2 
67. rcaprox 
68. rcadist 
69. lvx1: не используется
70. lvx2: не используется
71. lvx3: не используется
72. lvx4: не используется
73. lvf: не используется
