Информация о данных:

Каждый объект представляет собой пример руки, состоящей из пяти игральных карт, взятых из стандартной колоды 52. Каждая карта описывается с использованием двух атрибутов (масть и достоинство), в общей сложности 10 признаков. Есть признак class, который описывает "руку". Важен порядок карт, поэтому комбинаций 480.
