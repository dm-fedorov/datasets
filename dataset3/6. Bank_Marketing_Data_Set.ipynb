Bank Marketing Data Set 

Abstract: данные связаны с маркетинговыми кампаниями (телефонными звонками) португальского банковского учреждения. Цель классификации - предсказать, оформит ли клиент на срочный депозит (переменная y).

Характеристика данных: многомерный

Размер дата сета: 45211

Сфера: Бизнес

Характеристики признаков: Вещественные числа

Количество признаков: 17

Дата создания: 2012-02-14

Цель: Классификация

Пропущенные значения: N/A

Number of Web Hits: 809905

Информация о данных:

Данные связаны с маркетинговыми кампаниями банковского учреждения. Маркетинговые кампании были основаны на телефонных звонках. Часто требовалось более одного контакта с одним и тем же клиентом, чтобы клиент сделал срочный депозит ("да" или "нет").

1) bank-additional-full.csv со всеми примерами (41188) и 20 признаками, упорядоченными по дате (с мая 2008 года по ноябрь 2010 года), очень близко к анализированным данным [Moro et al., 2014]
2) bank-additional.csv  с 10% примеров (4119), случайно выбранных из данных 1) и 20 признаками.
3) bank-full.csv со всеми примерами и 17 признаками, упорядоченными по дате (более старая версия этого набора данных с меньшим количеством признаков). 
4) bank.csv с 10% примеров и 17 признакми, случайно выбранных из 3) (более старая версия этого набора данных с меньшим количеством признаков). 

Самые маленькие наборы данных предоставляются для тестирования более требовательных к вычислениям алгоритмов машинного обучения (например, SVM). 

Цель классификации - предсказать, оформит ли клиент (Да/нет) на срочный депозит (переменная y).

Описание данных:

Данные клиента банка:
- аge – возраст, количественный признак;
- job – тип работы, номинальный признак;
- marital – семейное положение, номинальный признак;
- education – образование, номинальный признак;
- default – имеет ли кредит, номинальный признак;
- housing – имеет ли кредит на жилье, номинальный признак;
- loan – имеет ли личные займы, номинальный признак.

Данные, связанные с последним контактом текущей кампании:
- contact – тип связи, номинальный признак;
- month – месяц последнего контакта с клиентом, номинальный признак;
- day_of_week – день недели последнего контакта с клиентом, номинальный признак.

Другие признаки:
- campaign – количество контактов, выполненных во время этой кампании и для этого клиента, количественный признак;
- pdays – количество дней, прошедших после последнего контакта с клиентом из предыдущей кампании, количественный признак (999 означает, что с клиентом ранее не связывались);
- previous – количество контактов, выполненных до этой кампании и для этого клиента, количественный признак;
- poutcome – результат предыдущей маркетинговой кампании, номинальный признак.

Признаки социально-экономического характера:
- emp.var.rate – уровень вариации занятости (квартальный показатель), количественный признак;
- cons.price.idx – индекс потребительских цен (месячный показатель), количественный признак;
- cons.conf.idx – индекс доверия потребителей (месячный показатель), количественный признак;
- euribor3m – ставка межбанковского кредитования (дневной показатель), количественный признак;
- nr.employed – количество работников (квартальный показатель), количественный признак.

Целевая переменная: y – подписал ли клиент срочный депозит, бинарный признак
