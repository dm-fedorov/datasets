```python
1. Title: Car Evaluation Database

2. Sources:
   (a) Creator: Marko Bohanec
   (b) Donors: Marko Bohanec   (marko.bohanec@ijs.si)
               Blaz Zupan      (blaz.zupan@ijs.si)
   (c) Date: June, 1997

3. Past Usage:

   The hierarchical decision model, from which this dataset is
   derived, was first presented in 

   M. Bohanec and V. Rajkovic: Knowledge acquisition and explanation for
   multi-attribute decision making. In 8th Intl Workshop on Expert
   Systems and their Applications, Avignon, France. pages 59-78, 1988.

   Within machine-learning, this dataset was used for the evaluation
   of HINT (Hierarchy INduction Tool), which was proved to be able to
   completely reconstruct the original hierarchical model. This,
   together with a comparison with C4.5, is presented in

   B. Zupan, M. Bohanec, I. Bratko, J. Demsar: Machine learning by
   function decomposition. ICML-97, Nashville, TN. 1997 (to appear)

4. Полезная информация:

   Car Evaluation Database was derived from a simple hierarchical
   decision model originally developed for the demonstration of DEX
   (M. Bohanec, V. Rajkovic: Expert system for decision
   making. Sistemica 1(1), pp. 145-157, 1990.). The model evaluates
   cars according to the following concept structure:

    	CAR - Автомобильная приемлемость;
   	PRICE - общая цена;
   	Buying – цена покупки;
   	maint - цена обслуживания;
   	TECH – технические характеристики;
   	COMFORT – комфорт;
   	doors - количество дверей;
   	persons - людей с точки зрения лиц для перевозки;
   	lug_boot - размер багажного отделения;
   	safety – примерная безопасность машины.

    Входные атрибуты печатаются в нижнем регистре. Помимо цели
    Концепция (CAR), модель включает в себя три промежуточные концепции:
   PRICE, TECH, COMFORT. Каждая концепция в оригинальной модели
    связанных с его потомками нижнего уровня набором примеров.
    База данных оценки автомобилей содержит примеры с
    структурированной информация, т. е. напрямую связанные с CAR
    атрибуты: покупка, обслуживание, двери, персоны, lug_boot, безопасность.

    Из-за известной базовой концепции эта база данных может быть
    особенно полезно для проверки конструктивной индукции и
    методы обнаружения структур.   

5. Количество наблюдений: 1728

6. Количество переменных: 6

7. Значения переменных:

   buying       очень высокая, высокая, средняя, низкая;
   maint        очень высокая, высокая, средняя, низкая;
   doors        2, 3, 4, 5 и более;
   persons      2, 4, более;
   lug_boot     маленькая, средняя, большая;
   safety       низкая, средняя, высокая.

8. Пропущенные значения: отсутствуют.

9. Распределение классов (количество наблюдений на класс)

   Класс	   Кол-во       Кол-во[%]
   ------------------------------------
   Не точные 		1210  	(70.023 %) 
   точные               384    (22.222 %) 
   хорошие        	69    	( 3.993 %) 
   очень хорошие      65  	( 3.762 %) 

```
