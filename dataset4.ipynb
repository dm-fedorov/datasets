Название: <b>Мониторинг состояния гидравлических систем</b>

Набор данных был получен экспериментально с помощью гидравлической испытательной установки. Эта  установка состоит из первичного и вторичного контура охлаждения-фильтрации, которые соединены через масляный бак. Система циклически повторяет процесс нагрузки (продолжительностью 60 секунд) и измеряет величины, такие как давление, объем потока и температуру, в то время как состояние четырех гидравлических компонентов (охладитель, клапан, насос и аккумулятор) варьируется.

Набор данных содержит необработанные данные датчика (то есть без выделения признаков), которые структурированы в виде матриц со строками, представляющими циклы, и столбцами точек в цикле. Используемые датчики:

Столбцы:

1: Состояние кулера:
3 - близко к полному отказу
20 - сниженная эффективность
100 - полная эффективность

2: Состояние клапана: 100 - оптимальное поведение переключения;
90 - небольшое отставание;
80 - серьезное отставание;
73 - близко к полному провалу;

3: Внутренняя утечка насоса:
0 - нет утечки;
1 - слабая утечка;
2 - сильная утечка;

4: Гидравлический аккумулятор:
130 - оптимальное давление;
115 - немного пониженное давление;
100 - сильно пониженное давление;
90 - близко к полному провалу;

5: Индикатор стабильности условий:
0 - условия были стабильными;
