Пример вызова информации о профессиональных игроках

```python
pro = requests.get("https://api.opendota.com/api/proPlayers").json()
```

Здесь лежит основная информация о профессиональных игроках игры

```python
pro
```

Попробуем достать ссылку на автарку какого- нибудь игрока, например Jerry

```python
for i in pro:
    if i['name'] == 'Jerry':
        j = i
```
