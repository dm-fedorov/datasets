Api для рaботы c Itunes Search

```python
import pandas as pd
import requests
```

Запрос на данные обо всех альбомах Джексона, название альбома, дата релиза, количество треков, жанр, и цена

```python
data = pd.DataFrame(requests.get('https://itunes.apple.com/search?term=michael+jacksonl&entity=album').json()['results'])
```

```python
db = data[['collectionName','releaseDate','trackCount','primaryGenreName','collectionPrice','currency']]
db
```
