```python
import requests
import pandas as pd
```

Пример получения прогноза погоды Лондона на 25.02.2019

Для начала узнаем ID города

```python
ident = requests.get("https://www.metaweather.com/api/location/search/?query=moscow").json()[0]["woeid"]
```

Теперь создаем запрос по данным о погоде Лондона на нужную дату

Данные на сайте собираются из разных источников. 
Можно посмотреть такие параметры, как атмосферное давление, максимальная и минимальная температура,скорость ветра, влажность, видимость и вероятность(предсказуемость) данного прогнозам.

```python
pd.DataFrame(requests.get("https://www.metaweather.com/api/location/"+ str(ident)+"/2019/2/25/").json())
```
