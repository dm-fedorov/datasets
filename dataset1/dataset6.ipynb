Название: <b>Популярность новостей в нескольких социальных сетях</b>

Большой набор данных о новостях и их соответствующих отзывов в социальных сетях на нескольких платформах: Facebook, Google+ и LinkedIn.

Собранные данные относятся к периоду 8 месяцев, между ноябрем 2015 года и июлем 2016 года, что составляет около 100 000 новостей по четырем различным темам: экономика, microsoft, Обама и Палестина. 
Этот набор данных предназначен для оценочных сравнений в задачах прогнозной аналитики, хотя и позволяет выполнять задачи в других областях исследований, таких как обнаружение и отслеживание тем, анализ тональности в кратком тексте, обнаружение первых новостей или рекомендация новостей. 

Более подробная информация о процессе создания набора данных приведена в статье, упомянутой в разделе "соответствующие документы". 

.R файл предоставлен, чтобы обеспечить простой импорт набора данных.

Атрибуты данных о новостях:

IDLink (числовой): уникальный идентификатор новостей<br> 
Title (string): заголовок новости cогласно официальным источникам в СМИ<br> 
Headline (string): заголовок новости cогласно официальным источникам в СМИ<br> 
Источник (строка): оригинальное новостное издание, опубликовавшее новость 
Тема (строка): тема запроса, используемая для получения элементов в официальных источниках мультимедиа<br> 
PublishDate (timestamp): дата и время публикации новостей <br>
SentimentTitle (numeric): оценка тональности текста в заголовке новостей <br>
SentimentHeadline (numeric): оценка тональности текста в заголовке новостей <br>
Facebook (numeric): конечное значение популярности новостей по данным источника в социальных сетях Facebook <br>
GooglePlus (numeric): конечное значение популярности новостей по данным источника в социальных сетях Google+ <br>
В LinkedIn (числовое): конечное значение популярности новостей по социальным СМИ источник в LinkedIn <br>

Атрибуты данных об отзывах в социальных сетях:

IDLink (числовой): уникальный идентификатор новостей <br>
TS1 (числовой): уровень популярности во временном срезе 1 (0-20 минут после публикации) <br>
TS2 (числовой): уровень популярности во временном срезе 2 (20-40 минут после публикации) <br>
TS... (числовой): уровень популярности во временном срезе ... <br>
TS144 (числовой): окончательный уровень популярности через 2 дня после публикации<br>
