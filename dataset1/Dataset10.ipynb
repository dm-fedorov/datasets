Название:<b>Контейнерный крановый контроллер</b>
    
Два прогнозирующих атрибута (скорость и угол) и один целевой атрибут (мощность).
Контейнерный кран выполняет функцию транспортировки контейнеров из одной точки в другую. Сложность этой задачи заключается в том, что контейнер соединен с мостовым краном с помощью кабелей, которые создают угол открытия во время транспортировки контейнера, что мешает работе на высоких скоростях из-за колебаний, возникающих в конечной точке, которые могут вызвать несчастные случаи.

Признаки:

- Скорость перемещения Контейнерного крана: низкая, средняя и высокая (низкая: 1, 2, 3; средняя: 6, 7, 8; высокая: 9, 10).
- Угол: большой отрицательный угол, маленький отрицательный угол, угол ноль, маленький положительный угол и большой положительный угол.
- Мощность: низкая, средняя и высокая (низкая: 0,3; средняя: 0,5; высокая: 0,7).
