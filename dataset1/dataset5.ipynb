Ссылка на датасет: https://archive.ics.uci.edu/ml/datasets/Sports+articles+for+objectivity+analysis

Название: <b>Спортивные статьи для объективного анализа</b>

1000 спортивных статей были помечены с помощью Amazon Mechanical Turk как объективные или субъективные. Предоставляются необработанные тексты, извлеченные функции и URL-адреса, из которых были извлечены статьи.

Атрибуты:

TextID - имя текстового файла <br>
URL - Ссылка на статью <br>
Label - метка статьи, объективная или субъективная <br>
totalWordsCount - общее количество слов в статье <br>
semanticobjscore - частота слов с объективной оценкой в SENTIWORDNET  <br>
semanticsubjscore - частота слов с субъективной оценкой в SENTIWORDNET <br>
CC - частота координационных союзов<br>
CD - частота цифр и количественных слов<br>
DT - частота определителей <br>
EX -  частота жизненных слов <br>
FW - частота иностранных слов <br>
INs - частота подчиненных предлогов или союзов<br>
JJ -  частота порядковых прилагательных или цифр <br>
JJR - частота сравнительных прилагательных <br>
JJS - частота превосходных прилагательных <br>
LS - частота элементов списка <br>
MD - частота модальных вспомогательных частей речи <br>
NN - частота исключительных нарицательных существительных<br>
NNP -  частота сингулярных собственных существительных <br>
NNPS - частота множественного числа собственных имен<br>
NNS - частота множественного числа существительных <br>
PDT - частота  определителей <br>
POS - частота родительных падежей<br>
PRP - частота личных местоимений <br>
PRP2 - частота притяжательных местоимений <br>
RB - частота наречий <br>
RBR - частота сравнительных наречий <br>
RBS - частота превосходных наречий <br>
RP - частота частиц <br>
SYM - частота символов  <br>
TOs - частота  "to" как предлога<br>
UH - частота междометий <br>
VB - частота глаголов базовой формы <br>
VBD - частота глаголов прошедшего времени <br>
VBG - частота настоящего причастия или глаголов герундия <br>
VBN - частота глаголов причастия прошедшего времени <br>
VBP - частота глаголов в настоящем времени с множественным числом субъектов 3-го лица <br>
VBZ - частота глаголов в настоящем времени с особыми субъектами от 3-го лица <br>
WDT - частота  определителей WH (what, who, which) <br>
WP - частота WH-местоимений <br>
WP$ - частота притяжательных WH-местоимений <br>
WRB - частота  WH-наречий <br>
baseform - частота  инфинитива глаголов <br>
Quotes - частота пар котировок во всей статье <br>
questionmarks - частота вопросительных знаков в статье <br>
exclamationmarks - частота восклицательных знаков в статье <br>
fullstops - частота точек <br>
commas - частота запятых <br>
semicolon - частота точек с запятой <br>
colon - частота двоеточий <br>
ellipsis - частота многоточия <br>
pronouns1st - частота местоимений первого лица (личные и притяжательные) <br>
pronouns2nd - частота местоимений второго лица (личные и притяжательные) <br>
pronouns3nd - частота местоимений третьего лица (личные и притяжательные) <br>
compsupadjadv - частота сравнительных и превосходных прилагательных и наречий <br>
past -  частота глаголов прошедшего времени с местоимениями от 1-го и 2-го лица <br>
imperative - частота повелительных глаголов <br>
present3rd - частота глаголов в настоящем времени с местоимениями от 3-го лица <br>
present1st2nd - частота глаголов в настоящем времени с местоимениями от 1-го и 2-го лица <br>
sentence1st - класс первого предложения<br>
sentencelast - класс последнего предложения <br>
txtcomplexity - оценка сложности текста <br>
