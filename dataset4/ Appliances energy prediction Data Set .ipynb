# <center>Прогнозирования энергопотребления приборов (Appliances energy prediction Data Set).                                       


### Описание:

Экспериментальные данные используемые для создания регрессионных моделей энергопотребления приборов в здании с низким потреблением энергии.
Набор данных составляет 10 минут собранных в течении примерно 4,5 месяцев. Температура и влажность в помещении контролировались с помощью беспроводной сенсорной сети ZigBee. Каждый беспроводной узел передавал температуру и влажность в течение 3,3 мин. Затем беспроводные данные были усреднены за 10 минут. Данные об энергии регистрировались каждые 10 минут с помощью счетчиков энергии m-bus. Погода с ближайшей метеостанции аэропорта (аэропорт Кьеврес, Бельгия) была загружена из общедоступного набора данных из Reliable Prognosis (rp5.ru) и объединена с наборами экспериментальных данных с использованием столбца даты и времени. Две случайные величины были включены в набор данных для тестирования регрессионных моделей и отфильтровывать не прогнозирующие атрибуты (параметры). 

Количество наблюдений: 19735

Количество атрибутов: 29

Пропусков в данных нет.

Для получения дополнительной информации о доме, сборе данных, сценариях R и рисунках, пожалуйста, обратитесь к статье и к следующему репозиторию github: https://github.com/LuisM78/Appliances-energy-prediction-data
