# <center> Отзывы на научные работы (Paper Reviews Data Set).

### Описание:

Набор данных для анализа содержит обзоры научных работ международной конференции
по вычислительной технике и информатике. Необходимо предсказать мнение на научную работу   или оценку её обзора.

### Информация о наборе данных:

Набор данных состоит из отзывов на научные работы, отправленных на международную
конференцию в основном на испанском языке (некоторые на английском языке). В 
общей сложности N = 405 экземпляров оцениваются по 5-балльной шкале («-2»: очень
отрицательный, «-1»: отрицательный, «0»: нейтральный, «1»: положительный, «2»:
очень положительный), выражая мнение рецензента о научной работе. Распределение 
исходных баллов является более равномерным по сравнению с пересмотренными 
баллами. Предполагается, что это различие происходит из-за несоответствия между
способом оценки статьи и тем, как рецензия написана первоначальным рецензентом. 

**Набор данных хранится в формате JSON, структура выглядит следующим образом: **

Paper: { 

Временной интервал и ID научной статьи, каждая научная работа содержит несколько научных отзывов. Рецензент имеет свой собственный ID, текст рецензии, примечания (которые могут быть пустыми), язык рецензии, его оценку.
Некоторая актуальная статистика (исключая рецензии на английском и пустые рецензии): 
- Количество слов: 
Min: 3 Max: 530 Avg: 88.64 Stdev: 69.76 
- Количество предложений: 
Min: 1 Max: 47 Avg: 8.91 Stdev: 7.54


### Информация об атрибутах:

1. Интервал времени (datetime): дата, связанная с годом конференции, которая, в свою очередь, соответствует времени написания обзора. Этот набор данных включает четыре года обзоров конференций. 
2. ID научной работы (integer) этот номер идентифицирует каждую работу с данной конференции. Набор данных имеет 172 различных документа. 
3. Предварительное решение (label): Предварительное решение о принятии или отклонении доклада, принятого комитетом конференции. 
4. ID рецензии (integer: идентификатор серийного номера для каждого рецензирования как коррелятивный по отношению к каждому отдельному документу. (Например, второй обзор некоторого реферата будет соответствовать числу $ 2 $). Набор данных содержит в общей сложности 405 рецензий. У большинства работ есть 2 отзыва. 
5. Текст (text): комментарии и подробный обзор статьи. Это читают авторы и редакция конференции. Редакция определяет, должна ли статья быть опубликована или нет, в зависимости от рецензий. Есть $ 6 $ экземпляров пустых отзывов. 
6. Замечания (text): Дополнительные комментарии, которые могут быть прочитаны только редакционной комиссией конференции. Это используется в сочетании с предыдущим атрибутом, чтобы определить, должна ли статья быть опубликована. Это необязательный атрибут. Всякий раз, когда это возможно, он объединяется в конце основной части обзора. В некоторых обзорах нет замечаний, это обозначается пустой строкой. 
7. Язык (text): язык, соответствующий обзору (может быть английский или испанский). В этом случае большинство обзоров составлено на испанском языке, а количество обзоров на английском языке составляет всего $ 17 $. 
8. Ориентация (intege от -2 до 2): обзорная классификация, определенная авторами данного исследования, согласно ранее описанной 5-балльной шкале, полученная путем систематического суждения авторов каждого обзора. Этот атрибут представляет субъективное восприятие каждого обзора (то есть, насколько отрицательный или положительный отзыв воспринимается, когда кто-то его читает). 
9. Оценка (intege от -2 до 2): просмотрите классификацию, определенную рецензентом, по ранее описанной 5-балльной шкале. Этот атрибут представляет реальную оценку, данную бумаге, как определено рецензентами. 
10. Достоверность (intege от 1 до 5): значение, описывающее достоверность рецензента, более высокое значение означает большую достоверность, а более низкое значение указывает на меньшую достоверность.
