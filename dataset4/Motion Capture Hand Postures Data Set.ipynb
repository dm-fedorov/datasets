# <center> Motion Capture позы рук (Motion Capture Hand Postures Data Set). 

### Описание:

Система камер видеонаблюдения Vicon использовалась для записи 12 пользователей, выполняющих 5 положений рук с маркерами, прикрепленными к левой перчатке. 
Маркеры на тыльной стороне перчатки использовался для создания локальной системы координат для руки, и 11 других маркеров были прикреплены к большому пальцу и пальцам перчатки. 3 маркера были прикреплены к большому пальцу, один над миниатюрой, а два других - на суставах пальцев. 2 маркера были прикреплены к каждому пальцу, один над ногтем, а другой на суставе между проксимальной и средней фалангой. 
11 маркеров, не являющихся частью жесткого шаблона, были немаркированными; их позиции явно не отслеживались. Следовательно, нет априорного соответствия между маркерами двух данных записей. Кроме того, из-за разрешения объема захвата и самопокрытия из-за ориентации и конфигурации руки и пальцев во многих записях отсутствуют маркеры. Посторонние маркеры были также возможны из-за артефактов в процессе реконструкции / записи маркеров программного обеспечения Vicon и других объектов в объеме захвата. В результате количество видимых маркеров в записи значительно варьировалось. 
Представленные здесь данные уже частично обработаны. Сначала все маркеры были преобразованы в локальную систему координат содержащей их записи. Во-вторых, каждый трансформированный маркер с нормой более 200 миллиметров был обрезан. Наконец, все записи, содержащие менее 3 маркеров, были удалены. Обработанные данные имеют не более 12 маркеров на запись и не менее 3. Для получения дополнительной информации см. «Информация об атрибутах». 
Из-за способа сбора данных вполне вероятно, что для данной записи и пользователя существует почти дублирующийся запись, исходящая от одного и того же пользователя. Поэтому мы рекомендуем оценивать алгоритмы классификации по принципу «оставь одного пользователя», в котором каждый пользователь итеративно не учится и используется в качестве тестового набора. Затем проверяется обобщение алгоритма для новых пользователей. Атрибут «Пользователь» предоставляется для размещения этой стратегии. 
Этот набор данных может использоваться для различных задач, наиболее очевидной из которых является распознавание позы по классификации. Можно также попытаться идентифицировать пользователя. Альтернативно, можно выполнить кластеризацию (ограниченную или неограниченную), чтобы обнаружить распределения маркеров, либо в качестве попытки предсказать идентичность маркеров, либо получить статистические описания визуализации положений. 
Случайным образом отбирали без замены постоянное количество (например, 75) записей на класс на пользователя для балансировки классов. Из-за разрешения и окклюзии пропущенные значения являются общими.


### Информация об атрибутах:

Данные предоставляются в виде файла CSV. Заголовок содержит имя каждого атрибута. Исходная фиктивная запись, состоящая полностью из 0, должна игнорироваться. Вопросительный знак '?' используется для указания отсутствующего значения. Запись соответствует одному моменту или кадру, записанному системой камеры. 

Количество отзывов: 78095

Количество атрибутов: 38

Пропуски есть в данных.

'Class' - Integer. Идентификатор класса данной записи. Диапазоны от 1 до 5 с 1 = кулак (с большим пальцем), 2 = стоп (рука плоская), 3 = точка 1 (точка с указательным пальцем), 4 = точка 2 (точка с указательным и средним пальцами), 5 = захват (пальцы свернувшись, как бы схватить). 

'User' - Integer. Идентификатор пользователя, который внес запись. Нет значения, кроме как в качестве идентификатора. 

'Xi' - Real. Координата x положения i-го немаркированного маркера «i» колеблется от 0 до 11. 

'Yi' - Real. Y-координата i-й позиции немаркированного маркера. «i» колеблется от 0 до 11. 

'Zi' - Real. Координата z позиции i-го немаркированного маркера. «i» находится в диапазоне от 0 до 11. 

Каждая запись является набором. I-й маркер данной записи не обязательно соответствует i-му маркеру другой записи. Можно случайным образом переставить видимые (то есть не пропущенные) маркеры данной записи без изменения набора, который представляет запись. Для удобства все видимые маркеры данной записи имеют более низкий индекс, чем любой отсутствующий маркер. Классу не гарантируется наличие даже одной записи со всеми видимыми маркерами.
