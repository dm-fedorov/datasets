# <center>Рисунки спиралей при болезни Паркинсона с использованием цифрового графического изображения (Parkinson Disease Spiral Drawings Using Digitized Graphics Tablet Data Set ).

### Описание:

Набор данных состоит из 62 человека PWP (Люди с паркинсонизмом) и 15 здоровых людей, обратившихся в отделение неврологии медицинского факультета Cerrahpasa Стамбульского университета. Из всех предметов берут три типа записей почерка (статический спиральный тест (SST), динамический спиральный тест (DST) и тест стабильности на определенную точку (STCP)). Также почерки, принадлежащих PWP, включены в набор данных в виде изображения. Поэтому этот набор данных также можно использовать для регрессии. 
Набор данных для рукописного ввода был создан с использованием таблицы графики Wacom Cintiq 12WX (Hahne et al., 2009). Это в основном графический планшет и ЖК-монитор, объединенные в одно целое. Он позволяет отображать экран ПК на мониторе и взаимодействует только с цифровыми ручками. Специальное программное обеспечение было разработано для записи почерка и тестирования координации пациентов с PWP с использованием записей. Программное обеспечение использует API-функции устройства и было разработано на платформе C #, которая может быть запущена в системах Windows. Вы можете связаться с авторами, чтобы запросить упомянутое программное обеспечение [1]. 
В этом исследовании для сбора данных с помощью графического планшета были разработаны три различных типа тестов. Первым из них является Статический спиральный тест (SST), который часто используется для клинических исследований в литературе для различных целей, таких как определение двигательной активности (Wang et al., 2008), измерение тремора (Pullman, 1998) и диагностика PD (Saunders et al.., 2008). В этом тесте три графические спирали Архимеда появляются на графическом планшете с использованием программного обеспечения, и пациентов просили отследить одну и ту же спираль настолько, насколько они могут, используя цифровую ручку. Во время теста перечисленные выше функции и другие данные, указывающие пациента, записываются в базу данных. 
Второй тест - динамический спиральный тест (DST). В отличие от SST, архимедова спираль просто появляется и исчезает через определенные промежутки времени, другими словами, архимедова спираль мигает. Это заставляет пациента помнить шаблон и продолжать рисовать. Цель этого теста состоит в том, чтобы определить изменение производительности рисования пациента и времени паузы, поскольку в этом случае сложнее проследить архимедову спираль. В результате этого теста наблюдается, что большинство пациентов продолжали рисовать, но почти все они потеряли образец. 
Третий тест - это тест на устойчивость к определенной точке (STCP). В этом тесте в центре экрана есть определенная красная точка, и испытуемых просят держать цифровую ручку на точке, не касаясь экрана в течение определенного времени. Цель этого теста - определить стабильность руки пациента или уровень дрожания рук.

Файл данных содержит набор данных рукописного ввода. В этом файле есть две разные папки, которые содержат две разные группы образцов почерка, которые называются PWP (люди с болезнью Паркинсона) и Здоровые.
