## 1. Общая информация
WESAD является общедоступным набором данных для носимых стрессов и обнаружения аффектов. Этот мультимодальный набор данных содержит физиологические данные и данные о движении, записанные как на запястье, так и на нагрудном устройстве, 15 человек во время лабораторного исследования. Включены следующие модальности датчика: пульс объема крови, электрокардиограмма, электродермальная активность, электромиограмма, дыхание, температура тела и ускорение по трем осям. Кроме того, набор данных устраняет разрыв между предыдущими лабораторными исследованиями стресса и эмоций, поскольку содержит три различных аффективных состояния (нейтральное состояние, стресс, развлечение). Кроме того, в набор данных включены самоотчеты субъектов, которые были получены с использованием нескольких установленных анкет.

## 2. Структура данных
Набор данных организован так, что у каждого испытуемого есть папка (SX, где X = идентификатор испытуемого). Каждая папка содержит следующие файлы: 
- SX_readme.txt: содержит информацию о испытуемом (SX) и информацию о сборе данных и качестве данных (если применимо)
- SX_quest.csv: содержит всю необходимую информацию для получения достоверной информации, включая расписание протокола для SX и ответы на вопросники для самоотчетов
- SX_respiban.txt: содержит данные с устройства RespiBAN
- SX_E4_Data.zip: содержит данные с устройства Empatica E4
- SX.pkl: содержит синхронизированные данные и метки

## 3. Испытуемые
17 человек участвовали в исследовании. Однако из-за неисправности датчика данные двух людей (S1 и S12) пришлось отбросить. Таким образом, соответствующие папки отсутствуют в WESAD. Информация по каждой теме может быть найдена в SX_readme.txt, в папке соответствующей темы.

## 4. Формат данных
Необработанные данные датчика были записаны на двух устройствах: нагрудное устройство (RespiBAN) и наручное устройство (Empatica E4). Метки протокола исследования синхронизируются с необработанными данными RespiBAN (то же время начала). Однако данные RespiBAN и Empatica E4 необходимо синхронизировать вручную. Испытуемые выполняли жест двойного постукивания недоминирующей рукой (где они носили E4) на груди. Результирующий характерный рисунок в сигнале ускорения может использоваться для синхронизации данных двух устройств. Кроме того, набор данных также включает файл SX.pkl, который включает в себя синхронизированные необработанные данные датчика и метки.

### 4.1 Данные с устройства RespiBAN
Все сигналы были выбраны с частотой 700 Гц. Необработанные данные содержатся в SX_respiban.txt. Здесь 10 столбцов. Первый столбец: порядковый номер строки. Второй столбец: игнорировать. Столбцы 3-10: необработанные данные 8 сенсорных каналов. Порядок каналов определяется в заголовке. Записи «XYZ» относятся к 3-канальному акселерометру (таким образом, данные ускорения представлены в 3 столбцах). Чтобы преобразовать необработанные значения датчика в единицы СИ, каждый канал необходимо преобразовать на основе приведенных ниже формул (сигнал содержит необработанные значения датчика, vcc = 3, chan_bit = 2 ^ 16).
#### Атрибуты
- электрокардиограмма (ЭКГ)
- электродермальная активность (ЭДА)
- электромиограмма (ЭМГ)
- дыхание
- температура тела
- трехосевое ускорение


### 4.2 Данные с устройства Empatica E4
Устройство Empatica E4 носили на недоминантном запястье испытуемых. Частота дискретизации разных датчиков была различной. Необработанные данные содержатся в SX_E4_Data.zip. После распаковки следующие файлы содержат производную информацию и поэтому должны игнорироваться в этом наборе данных: HR.csv, IBI.csv, tags.csv. Файл info.txt содержит некоторые сведения о содержимом папки. Необработанные данные с устройства E4 содержатся в следующих файлах (в каждом файле первая строка относится к глобальной временной метке канала датчика при запуске, вторая строка относится к частоте дискретизации канала датчика):
#### Атрибуты
- ACC.csv: данные с 3-х осевого датчика акселерометра. частота дискретизации 32 Гц. 3 столбца данных относятся к 3 каналам акселерометра. Данные предоставляются в единицах 1/64г.
- BVP.csv: пульс объема крови, частота дискретизации 64 Гц. Данные из фотоплетизмографа (PPG).
- EDA.csv: электродермальная активность, выборка с частотой 4 Гц. Данные представлены в мкСм.
- TEMP.csv: температура тела, выборка с частотой 4 Гц. Данные представлены в ° C.
