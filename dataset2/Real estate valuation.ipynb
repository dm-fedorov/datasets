### Общая информация: рыночные исторические данные об оценке недвижимости собранные в Синдиане, Нью-Тайбэй, Тайвань
### Информация об атрибутах: 
    Признаки:
        X1 the transaction date - дата транзакции (например: 2013.250=2013 Март, 2013.500=2013 Июнь и тд.)
        X2 house age - возраст дома (в годах)
        X3 distance to the nearest MRT station - расстояние до ближайшей станции метро (в метрах)
        X4 number of convenience stores - количество круглосуточных магазинов в жилом круге пешком (целое число)
        X5 latitude - географическая координата, широта. (в градусах)
        X6 longitude - географическая координата, долгота. (в градусах)
    Целевая переменная: 
        Y house price of unit area - цена дома за единицу площади (10000 новых тайваньских долларов / пинг, где пинг является местной единицей, 1 пинг = 3,3 метра в квадрате)
