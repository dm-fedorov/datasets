Название: <b>Нечестные интернет пользователи</b>

В распределенных в сети  вычислениях взаимодействующие пользователи не могут получить информацию о достоверности друг друга. Таким образом, недобросовестные пользователи могут действовать злонамеренно по отношению к другим. Предлагаемое решение позволяет оценить надежность каждого пользователя путем мониторинга поведения друг друга во время их взаимодействия в сети. Эти поведения представлены кортежами, включающими значимые параметры. Основанная на этих кортежах архитектура объединяет некоторые технологии на основе искусственного интеллекта для реализации системы принятия решений.
Кортежи таковы:
- eij - i-й пользователь, взаимодействующий с j-м пользователем.</br>
- EIDj - Идентификация j-го пользователя.</br> 
Значения целевого условия помечаются аннотацией в файле profile.txt (с разделителями табуляции). Номер строки представляет номер цикла.</br>
- CТ - Подсчет Доверия. Он используется для подсчета количества надежных транзакций (принадлежащих определенному контексту) после последней ненадежной транзакции.</br> 
- CU - Подсчет недоверbя. Он используется для подсчета количества ненадежных транзакций (принадлежащих конкретному контексту) после последней заслуживающей доверия транзакции.</br> 
- LT -  Последний раз. Он используется для учета даты, когда произошел последний опыт в определенном контексте.</br> 
- TC - контекст транзакций. Он используется для определения типа транзакции, такой как игра, электронная коммерция, социальная сеть и другие.</br>  
- TS - оценка доверия. Это оценка, которую объект дает другому объекту в конце каждого прямого взаимодействия.</br>     
Набор данных был получен с помощью симулятора Java, который реализовал предложенную архитектуру.
Он включает в себя данные для трех самых популярных типов атак, а именно:</br>    
- Счетная атака. Пользователь пытается завоевать хорошую репутацию, чередуя честное и нечестное поведение.
- Атака по времени. Пользователь снова пытается завоевать хорошую репутацию, чередуя честное и нечестное поведение, но действует в другое время.
- Контекстная атака. Пользоваптель пытается завоевать хорошую репутацию, действуя честно по отношению к одному типу транзакции и нечестно по отношению к другому.



Поскольку может возникнуть ситуация, когда у пользователей нет исторических данных (кортежей) для взаимодействия с другим, он может получать данные (кортежи) от сторонних организаций, которые ранее имели взаимодействие с запрашиваемым пользователем.
Тем не менее, надежность таких сторонних организаций (рекомендателей) также должна быть оценена. Действительно, они могут действовать посредством атак, таких как: заполнение бюллетеней для голосования (BS), плохое высказывание и случайное мнение (RO).
Изменение параметра TS для ряда строк в наборе данных и в соответствии с конкретной атакой позволяет получить различные наборы данных, полезные для оценки надежности рекомендателей.
В соответствии с этим также предоставляются следующие наборы данных:
- BM_x% .txt x - процент несправедливых рекомендаций, полученных атакой BM. Он варьируется от 10 до 50.
- BS_x% .txt x - это процент недобросовестных рекомендаций, полученных атакой BS. Он варьируется от 10 до 50.
- RO_x% .txt x - процент несправедливых рекомендаций, полученных атакой BM. Он варьируется от 10 до 50.

Информация об атрибутах:
-  1) CT {CT_range_1, CT_range_2, CT_range_3, CT_range_4} 
- 2) CU {CU_range_1, CU_range_2, CU_range_3, CU_range_4} 
- 3) LT {LT_range_1, LT_range_2, LT_range_3, LT_range_4} 
- 4) TC {sport, game, ECommerce, holiday} 
- 5) TS {trustworthy, untrustworthy}
