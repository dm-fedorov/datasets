Название: <b>Имитация падений и повседневных активностей человека</b>

17 испытуемых выполнили 20 падений и 16 повседневных активностей с 5 повторениями. На испытуемых было 6 датчиков, которые прикреплялись к их голове, груди, пояснице, запястью, бедру и лодыжке. Каждый из 5 датчиков включает в себя 3-х осевой акселерометр, гироскоп и магнитометр.

Количество наблюдений:  3060 (17 испытуемых × 5 повторений × 36 движений)

Данные сформированы таким образом:

№ папки	  Содержание</br>   
1XX Испытуемые мужчины</br>  
2XX	Испытуемые женщины</br>  
8XX	Повседневная активность</br>  
9XX	Падения</br>
